```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame (Replace this with your actual DataFrame)
df = pd.read_csv("/content/athlete_events.csv_silver.csv")  # Load your dataset

# Select only numerical columns
num_cols = df.select_dtypes(include=['number']).columns

# Set up the figure size
plt.figure(figsize=(15, 6 * len(num_cols)))

# Loop through each numerical column and plot distribution + boxplot
for i, col in enumerate(num_cols):
    plt.subplot(len(num_cols), 2, 2 * i + 1)
    sns.histplot(df[col], bins=30, kde=True, color="blue")
    plt.title(f"Distribution of {col}")

    plt.subplot(len(num_cols), 2, 2 * i + 2)
    sns.boxplot(x=df[col], color="red")
    plt.title(f"Box Plot of {col}")

plt.tight_layout()
plt.show()
```
